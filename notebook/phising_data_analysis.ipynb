```python
import numpy as np
import pandas as pd
from sklearn import *

from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from scipy.io.arff import loadarff #for load arff dataset format
```

```python
#Importing dataset from arff file
phising_dataset_raw = loadarff("/home/wirl-1424/dataset/phising_data/PhishingData.arff")
phising_dataset_raw
```

```python
#selectiing of the data and converting data into numpy array for flexibility in cleaning
phising_da = np.array(phising_dataset_raw[0])
phising_da
```

```python
#converting from numpy array info dataframe format using pandas
phising_df = pd.DataFrame(phising_da).apply(pd.to_numeric)
phising_df
```

```python
#take the main of dataset
phising_df.columns = [''] * len(phising_df.columns)
```

```python
#save dataset to csv
phising_df.to_csv('dataset.csv',header=False, index=False)
```

```python
#import the new dataset with the main data
training_data = np.genfromtxt('dataset.csv', delimiter=',',dtype=np.int32)
```

```python
#show the numpy array of dataset
training_data
```

```python
#identify the inputs(all of the attributes, except for the last one) and the ouptus (the last atrribute):
inputs = training_data[:,:-1]
outputs = training_data[:, -1]
```

```python
#training and testing processes
training_inputs = inputs[:2000]
training_outputs = outputs[:2000]
testing_inputs = inputs[2000:]
testing_outputs = outputs[2000:]
```

```python
#create the model
model = { "Logistic Regression":LogisticRegression(),
         "Decision Tree":tree.DecisionTreeClassifier(),
         "Random Forest":RandomForestClassifier(),
         "Linear SVM":svm.SVC(),
         "Gradient Booted":GradientBoostingClassifier(),
         "Multilayer Perceptron":MLPClassifier()
}
```

```python
results_algo = {}
for algorithm in model:
    classifier = model[algorithm]
    classifier.fit(training_inputs, training_outputs)
    predictions = classifier.predict(testing_inputs)
    accuracy = 100.0 * accuracy_score(testing_outputs, predictions)
    print("Algorithm "+str(algorithm)+", Accuracy: "+str(accuracy))
    results_algo[algorithm] = accuracy
    
#show the best algorithm
print("==============================================================")
best_algo = max(results_algo, key=results_algo.get)
#best_value = max(results_value, key=results_value.get)
print("Best results is : "+best_algo)
```
